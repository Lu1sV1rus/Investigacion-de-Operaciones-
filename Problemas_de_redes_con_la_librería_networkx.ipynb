<p><font color="tomato" size="5">&nbsp;&nbsp;Despues)</font></p>

<p><font color="tomato" size="4">&nbsp;&nbsp;&nbsp;&nbsp;
Creamos un grafo digido usando la libreria
</font><font size="4">nx.DiGraph</p></font>

<p><font color="tomato" size="4">&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
Lo guardamos en una varible </font><font size="4">G</p></font>

<p><font size="3">&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
Un grafo dirido contiene las flechas que reflejan el flujo en cada arco
</font></p>

<font color="tomato" size="4"><p>&nbsp;&nbsp;&nbsp;&nbsp;Agregamos los nodos y los arcos </font><font size="4">G.add_weighted_edges_from([("H","K",N)])</p></font>

<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
add_weighted_edges_from([]) nos permite agregar arcos al grafo de forma directa
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
"H" es el nombre del nodo "inicial"
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
"K" es el nombre del nodo "final"
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
"N" es el flujo que tendra el arco de "H" a "K"</p>

<p><font color="tomato" size="4">&nbsp;&nbsp;&nbsp;&nbsp;
Graficamos la Red con </font><font size="4">nx.draw_networkx(G, node_color="tomato")</p><font color="tomato" size="3"><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
nx.draw_networkx() nos permite dibujar la red
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
"G" es el nombre del grafo a dibujar
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
node_color="" nos permite cambiar el color de los nodos</p>

```python
G=nx.DiGraph()
G.add_weighted_edges_from([
  ("A","B",9),
  ("A","C",7),
  ("B","E",2),
  ("C","D",4),
  ("C","E",6),
  ("B","D",7),
  ("D","F",6),
  ("D","E",3),
  ("E","F",9)
  ])
nx.draw_networkx(G, node_color='tomato')
```

<p><font color="lime" size="5">&nbsp;&nbsp;Arbol de expansi√≥n m√≠nima)</font></p>
<p><font color="lime" size="4">&nbsp;&nbsp;&nbsp;&nbsp;
Usamos </font><font size="4">nx.minimum_spanning_tree(nx.to_undirected(G))</p>
<font size="3">
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
nx.minimum_spanning_tree() usamos el comando para crear un arbol de minima expancion
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
nx.to_undirected() transformamos a de G de un digrago a un grafo, para que minimum_spanning_tree pueda hacer el arbol de minima expancion
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
G es el nombre de la red que queremos graficar
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;

</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;

```python
# Arbol de expansi√≥n m√≠nima
S=nx.minimum_spanning_tree(nx.to_undirected(G))
print(S)
nx.draw_networkx(S,node_color='lime')
```

<p><font size="4">

<font size="3">
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
nx.shortest_path_length() usamos el comando para calcular la distancia de la ruta mas corta
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;

<p><font size="4">

```python
# Ruta m√°s corta
T=nx.shortest_path(G, "A", "F", weight='weight')
distancia = nx.shortest_path_length(G, "A", "F", weight='weight')

print(f"Ruta m√°s corta: {T}")
print(f"Distancia: {distancia}")

# Crear subgrafo con solo la ruta m√°s corta
ruta = nx.path_graph(T)
print(ruta)
nx.draw_networkx(ruta, node_color='skyblue')
```

<p><font color="violet" size="5">&nbsp;&nbsp;Flujo Maximo)</font></p>
<p><font color="violet" size="4">&nbsp;&nbsp;&nbsp;&nbsp;
Usamos </font><font size="4">nx.maximum_flow(G, "A", "F",capacity="weight")</p>
<font size="3">
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
nx.maximum_flow() usamos el comando para encontrar el flujo maximo
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
G es el nombre de la red que queremos graficar
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
"A" es el nombre del nodo "inicial"
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
"F" es el nombre del nodo "final"
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
weight es el peso que tiene cada arco
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
K es una lista que contienes el flujo y los nodos</p>
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
K[0] es el flujo de la red
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
K[1] una lista de los nodos
</p>


<p><font color="violet" size="4">
&nbsp;&nbsp;&nbsp;&nbsp;
La red de flujo maximo</font><font size="3">
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
for para iterar y recorrer K[1]
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
.items() para obtener la capacidad
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
finalmente los agregamos al grafo llamado X, con add_edge()
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
Graficamos la Red con nx.draw_networkx(X)</p>



```python
# Flujo m√°xima
K=nx.maximum_flow(G, "A", "F",capacity="weight")

#Creamos la red de flujo maximo
X=nx.DiGraph()
for u in K[1]:
    for v, flow in K[1][u].items():
        if flow > 0:
            X.add_edge(u,v,capacity=flow)

print(f"Red:{K}")
print(X)
print(f"El flujo m√°ximo: {K[0]}")

# Dibujar la red de flujo
nx.draw_networkx(X, node_color='violet')
```
